## 3. ***Data Wrangling***

### Data Wrangling Code

```python

# Import library
import pandas as pd

# Load datasets
titles = pd.read_csv('titles.csv')
credits = pd.read_csv('credits.csv')

# Merge datasets using common column 'id'
data = pd.merge(titles, credits, on='id', how='inner')

# -------------------------
# Handling Missing Values
# -------------------------

# Remove missing descriptions
data.dropna(subset=['description'], inplace=True)

# Fill categorical missing values
data['age_certification'].fillna('Unknown', inplace=True)
data['character'].fillna('unknown', inplace=True)

# Fill numerical missing values
data['imdb_score'].fillna(round(data['imdb_score'].mean(),1), inplace=True)

data['imdb_votes'].fillna(0, inplace=True)

data['tmdb_popularity'].fillna(round(data['tmdb_popularity'].mean(),2), inplace=True)

data['tmdb_score'].fillna(round(data['tmdb_score'].mean(),1), inplace=True)

# Fill seasons null values with 0
data['seasons'].fillna(0, inplace=True)

# Optional: Fill imdb_id missing values
data['imdb_id'].fillna('Unknown', inplace=True)

# -------------------------
# Remove Duplicates
# -------------------------
data.drop_duplicates(inplace=True)

# -------------------------
# Final Validation
# -------------------------

print("Dataset Shape:", data.shape)

print("\nMissing Values:\n")
print(data.isnull().sum())

print("\nDuplicate Records:", data.duplicated().sum())

print("\nData Ready for Analysis")
```

3. Missing values were treated using appropriate techniques:

    Duplicate records were checked and removed if found to maintain data quality.

4. Validated Data Types

    Column data types were checked to ensure numerical and categorical variables were suitable for analysis.

5. Prepared Dataset for Analysis

    The dataset was cleaned and transformed into an analysis-ready format for visualization and exploratory data analysis.

**Insights Found:**

**Multivariate Analysis**

Multivariate analysis examines relationships among three or more variables simultaneously. It helps identify interactions, patterns, and combined effects of multiple variables using advanced visualizations such as grouped charts, heatmaps, and multivariable comparisons.

#### Chart - 14 - Correlation Heatmap

**Correlation Analysis**

Correlation analysis measures the strength and direction of the relationship between numerical variables. It shows whether variables move together positively, negatively, or have little relationship, often using correlation matrices and heatmaps.

**Purpose:** To identify relationships between numerical variables and measure their dependency.

```python
#Full Numerical Correlation Matrix
import numpy as np
numeric_cols = data.select_dtypes(include=np.number)
plt.figure(figsize=(10,6))
sns.heatmap(numeric_cols.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

##### 1. Why did you pick the specific chart?

A heatmap is ideal for multivariate analysis because it shows relationships among multiple numerical variables in one chart. Instead of comparing two variables at a time, it gives a full relationship overview.


##### 2. What is/are the insight(s) found from the chart?

Understand relationships among all numerical variables.
